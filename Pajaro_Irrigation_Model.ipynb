### **1. Data Ingestion**
We begin by loading the raw data files.
*   **Water Data:** Extracted from district monthly reports.
*   **Weather Data:** Downloaded directly from the California Irrigation Management Information System (CIMIS).

```python
water_df = pd.read_csv('water_deliveries.csv')
water_df.head()
```

```python
weather_df = pd.read_csv('monthly_weather.csv')
weather_df.head()
```

```python
weather_df.columns
```

### **2. Data Cleaning & Integration**
Raw data is rarely ready for analysis. In this step, we:
1.  **Merge** the two datasets using the date as the common key.
2.  **Filter** out Quality Control flags and static metadata (Station ID, Name) to ensure the dataset is purely numerical for modeling.
3.  **Sort** the timeline chronologically to ensure accurate historical tracking.

```python
# 1. STANDARDIZE DATE FORMATS
# Water data uses "21-Jan" (Year-Month) -> Convert to datetime
water_df['Month Year'] = pd.to_datetime(water_df['Month Year'], format='%y-%b')

# Weather data uses "Jan-21" (Month-Year) -> Convert to datetime
weather_df['Month Year'] = pd.to_datetime(weather_df['Month Year'], format='%b-%y')

# 2. MERGE THE DATA
# Now that both columns are proper 'datetime' objects, we can merge them safely.
# an 'inner' merge means we only keep months that exist in BOTH files.
df_merged = pd.merge(water_df, weather_df, on='Month Year', how='inner')

# Verify the merge worked
print(f"Success! Merged dataset has {df_merged.shape[0]} months of data.")
df_merged.head()
```

```python
# 1. CREATE A CLEAN COPY
df_clean = df_merged.copy()

# 2. RENAME "Month Year" to "Date"
df_clean = df_clean.rename(columns={'Month Year': 'Date'})

# 3. DROP "QC" COLUMNS
# Identify columns with 'qc' in the name
qc_cols = [col for col in df_clean.columns if 'qc' in col.lower()]
df_clean = df_clean.drop(columns=qc_cols)

# 3.5 DROP STATIC TEXT COLUMNS (CRITICAL FIX)
# We drop Station Name and Region because they are text (which breaks the math)
# and they are the same for every row (so they provide no predictive info).
static_cols = ['Stn Id', 'Stn Name', 'CIMIS Region']
df_clean = df_clean.drop(columns=static_cols)

# 4. MOVE 'Date' TO THE FRONT
cols = ['Date'] + [c for c in df_clean.columns if c != 'Date']
df_clean = df_clean[cols]

# 5. CHECK THE RESULT
print(f"Final Clean Shape: {df_clean.shape}")
# We expect to see only numeric columns (plus Date) now
display(df_clean.head())
```

### **3.  (Correlation Analysis)**
Before building an AI model, we must verify that the data follows physical laws.
*   **Positive Correlation :** Variables that *increase* water demand (e.g., ETo, Solar Radiation, Soil Temp).
*   **Negative Correlation :** Variables that *decrease* water demand (e.g., Precipitation, Humidity).

```python
# 5. CORRELATION MATRIX (The "Insight" Step)
# This tells us which weather variables actually matter.
# We only care about correlations with 'Total Delivered (AF)'
numeric_df = df_clean.select_dtypes(include=['float64', 'int64'])
correlations = numeric_df.corr()['Total Delivered (AF)'].sort_values(ascending=False)

print("\n=== WHAT DRIVES WATER DEMAND? ===")
print(correlations)
```

### **4. Machine Learning: The Random Forest Model**
We are using a **Random Forest Regressor** to predict water demand.

**How it works:**
Instead of a single equation, the model builds a "committee" of 100 decision trees. Each tree looks at the weather conditions and votes on what the water demand should be. The final prediction is the average of these votes. This method is highly effective for handling the complex, non-linear relationship between weather and farmer behavior.

*   **Training Set:** 70% of the historical data (used to teach the model).
*   **Testing Set:** 30% of the data (hidden from the model to test its accuracy).

### **4b. Feature Engineering and Selection Strategy**
To build a robust predictive model, we carefully selected inputs based on physical causality and statistical correlation.

*   **Engineered Features:** We created **Net ETo** and define it as Total ETo - Total Precip in order to explicitly define water defecit for the model to see if it improves prediction compared to training on Precip and ETo as stand alone features
*   **Included Features:** We selected the newly created **Net ETo**, and the existing features **Solar Radiation**, and **Soil Temperature** because our correlation analysis confirmed these are the strongest external drivers of agricultural water demand.
*   **Excluded Variables:** We explicitly removed **Recycled Water** and **Blend Water** from the input set. These variables are constituent parts of the *Total Delivered* target. Including them would cause **"Target Leakage"**â€”a scenario where the model is inadvertently given the answer it is trying to predict, rendering the evaluation metrics invalid.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score

# 1. CALCULATE NET ET
# Physics: Plant Demand (ETo) minus Free Water (Rain)
df_clean['Net ET (in)'] = df_clean['Total ETo (in)'] - df_clean['Total Precip (in)']

# Optional: Clip negative values to 0
# If 5 inches of rain falls but ETo is only 2 inches, Net Demand is 0 (not -3).
# Uncomment the line below if you want to enforce this physical rule:
# df_clean['Net ET (in)'] = df_clean['Net ET (in)'].clip(lower=0)

# 2. DEFINE FEATURES AND TARGET
# We include our new "Net ET" but EXCLUDE the original ETo and Precip columns
# to prevent the model from getting confused by duplicate information.
features = [
    'Net ET (in)',
    'Avg Sol Rad (Ly/day)',
    'Avg Soil Temp (F)'
    # You can add 'Avg Wind Speed (mph)' here later if you want to test it
]

target = 'Total Delivered (AF)'

# 3. CREATE X and y
X = df_clean[features]
y = df_clean[target]

# 4. VERIFY
print("Feature Engineering Complete.")
print(f"Features Used: {features}")
print("\nFirst 5 Rows of X (Predictors):")
display(X.head())
```

```python
# 2. SPLIT DATA
# Since we have a small dataset (monthly), we will train on 80% and test on 20%
# random_state=42 ensures we get the same split every time we run this.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 3. TRAIN THE MODEL
# n_estimators=100 means "create 100 decision trees and average their answers"
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
# 4. MAKE PREDICTIONS
# We predict on the Test set (to see how it handles "new" data)
# AND the whole dataset (to visualize the full timeline)
y_pred_test = model.predict(X_test)
df_clean['Predicted Demand (AF)'] = model.predict(X)
```

## 4c. Validation Experiment: Limited Features vs. The "Kitchen Sink"

To ensure our feature selection was robust, we ran a side-by-side comparison between:
1.  **The Focused Model:** Uses only 3 physics-based features (`Net ET`, `Solar Radiation`, `Soil Temp`).
2.  **The Kitchen Sink Model:** Uses all 15 available numerical weather variables.

### **The Results**

| Metric | Focused Model (3 Features) | Kitchen Sink (15 Features) | Interpretation |
| :--- | :--- | :--- | :--- |
| **Training RÂ²** | 0.980 | 0.988 | The complex model "memorized" the training data slightly better. |
| **Testing RÂ²** | **0.890** | 0.887 | **CRITICAL:** The complex model performed *worse* on new data. |
| **Overfitting Gap** | 0.091 | 0.101 | The gap between training and testing widened with more features. |

### **Conclusion: The "Curse of Dimensionality"**
The experiment confirms that adding more variables **did not improve predictive power**. Instead, it introduced noise.

*   **The Kitchen Sink model** fell into the trap of overfittingâ€”it found random patterns in the training data that don't exist in the real world (hence the lower Test RÂ²).
*   **The Focused model** is superior because it achieves better accuracy with simpler inputs, making it more robust and easier to explain operationally.

**Decision:** We will proceed with the **Focused Model** for the final production pipeline.

```python
# 1. DEFINE TWO FEATURE SETS
# Set A: The "Physics-Based" Selection
features_focused = ['Net ET (in)', 'Avg Sol Rad (Ly/day)', 'Avg Soil Temp (F)']

# Set B: The "Kitchen Sink" (All numeric weather columns)
# We exclude the target and the cheat columns (Blend/Recycled)
all_cols = df_clean.select_dtypes(include=['float64', 'int64']).columns.tolist()
features_all = [c for c in all_cols if c not in ['Total Delivered (AF)', 'Blend Water (AF)', 'Recycled Water (AF)']]

# 2. TRAIN & EVALUATE BOTH MODELS
for name, feature_set in [("Focused Model", features_focused), ("Kitchen Sink Model", features_all)]:
    print(f"\n--- Training {name} ---")
    print(f"Features: {len(feature_set)}")

    # Define X and y
    X = df_clean[feature_set]
    y = df_clean['Total Delivered (AF)']

    # Split (Same random seed ensures fair comparison)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Train
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)

    # Predict
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)

    # Score
    r2_train = r2_score(y_train, train_pred)
    r2_test = r2_score(y_test, test_pred)

    print(f"Training RÂ²: {r2_train:.3f}")
    print(f"Testing  RÂ²: {r2_test:.3f}")
    print(f"Overfitting Gap: {r2_train - r2_test:.3f}")
```

```python
# --- 5. FINAL MODEL & VISUALIZATION ---

# 1. RETRAIN THE BEST MODEL (Just to be safe!)
# We stick to the 'Focused' features we proved were better
X = df_clean[features_focused]
y = df_clean['Total Delivered (AF)']

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Fit
final_model = RandomForestRegressor(n_estimators=100, random_state=42)
final_model.fit(X_train, y_train)

# Predict
y_pred_test = final_model.predict(X_test)

# 2. CALCULATE METRICS
mae = mean_absolute_error(y_test, y_pred_test)
r2 = r2_score(y_test, y_pred_test)

print(f"=== MODEL REPORT ===")
print(f"Mean Absolute Error: {mae:.0f} AF")
print(f"(On average, the model is off by about {mae:.0f} Acre-Feet per month)")
print(f"RÂ² Score: {r2:.2f} (1.0 is perfect)")
```

```python
# --- HYPOTHESIS TEST: UNCOUPLING PRECIP & ETO ---

# 1. Define features with SEPARATE Rain and ETo
# We keep Solar and Soil Temp, but break Net ET back into its parts
features_uncoupled = ['Total ETo (in)', 'Total Precip (in)', 'Avg Sol Rad (Ly/day)', 'Avg Soil Temp (F)']

# 2. Retrain the model
X = df_clean[features_uncoupled]
y = df_clean['Total Delivered (AF)']

# Use same seed (42) for fair comparison
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model_uncoupled = RandomForestRegressor(n_estimators=100, random_state=42)
model_uncoupled.fit(X_train, y_train)

# 3. Check the results
r2_uncoupled = r2_score(y_test, model_uncoupled.predict(X_test))
mae_uncoupled = mean_absolute_error(y_test, model_uncoupled.predict(X_test))

print(f"=== TEST RESULTS (Separate Precip & ETo) ===")
print(f"RÂ² Score: {r2_uncoupled:.3f} (Compare to Focused Model: 0.89)")
print(f"MAE: {mae_uncoupled:.0f} AF")

# 4. Feature Importance (Did Precip pull its weight?)
importances = pd.Series(model_uncoupled.feature_importances_, index=features_uncoupled).sort_values(ascending=False)
print("\nNew Feature Importances:")
print(importances)
```

### **8. Conclusion & Next Steps**

**Results Summary**
This prototype demonstrates that **Weather Data alone accounts for approximately 89% of the variance** in the district's monthly water demand ($R^2=0.89$). This strong correlation supports the feasibility of a weather-based forecasting tool.

Our Feature Importance analysis revealed a distinct physical hierarchy:
1.  **Soil Temperature (~48%)** acts as the **"Seasonal Regime Indicator."** It identifies the Cool/Wet Season. In this window, irrigation is minimal due to a combination of **low biological activity** (small/dormant plants) and **precipitation** (captured implicitly because cold months are usually wet months).
2.  **Atmospheric Demand (~52%)** (Solar Radiation + Net ET) acts as the **"Intensity Adjuster."** Once the season is active, these factors determine the precise volume of water needed based on heat, evaporation, and specific rainfall deficits (`Net ET`).

**Potential Improvements (Operational Roadmap)**
To transition this tool from a planning prototype to an operational dashboard, we recommend the following enhancements:

1.  **Daily Granularity:** Transition from monthly aggregates to daily SCADA meter readings. Monthly data hides short-term "demand spikes" that strain infrastructure; daily data will expose them.
2.  **System Inertia (Autoregressive Features):** Incorporate *Past Water Deliveries* (e.g., Demand Yesterday, Demand Last Week) as inputs.
    *   *Reasoning:* Irrigation behavior has "momentum." If a grower turns on a 3-day drip cycle, they are likely to continue tomorrow regardless of a small weather change. Combining **System History** with **Weather Forecasts** is the standard for high-accuracy operational modeling.
3.  **Production Pipeline:** Automate the ingestion of live CIMIS weather API data to generate a rolling 7-day demand outlook for pump operators.
